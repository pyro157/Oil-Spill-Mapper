```python
df1 = pd.read_csv(data1_path)
df2 = pd.read_csv(data2_path)
df3 = pd.read_csv(data3_path)
#df4 = pd.read_csv(data4_path)
df5 = pd.read_csv(data5_path)
```

```python
df = pd.concat([df1,df2,df3.iloc[13:,:],df5], axis =0)
df = df.reset_index(drop=True)
```

```python
df
```

```python
# Set pandas display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Select the "WELL NAME" and "WELL NUMBER" columns
well_name = df["WELL NAME"]
well_number = df["WELL NUMBER"]

# Concatenate the two columns side by side
side_by_side_df = pd.concat([well_name, well_number], axis=1)

# Print the side by side DataFrame
print(side_by_side_df)

# Reset the display options to their default values if needed
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

This code will fill missing values in the "WELL NAME" column with the values from the "WELL NUMBER" column. If "WELL NAME" has information, it will be kept as is. If it doesn't have information, it will be replaced with the "WELL NUMBER" information.

```python
# Assuming df is your DataFrame with "WELL NAME" and "WELL NUMBER" columns
df["WELL NAME"] = df["WELL NAME"].fillna(df["WELL NUMBER"])
```

```python
# Set pandas display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print(df["WELL NAME"])

# Reset the display options to their default values if needed
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```
